## Sigmoid function 

#### Create a sigmoid function with following parameters
- mid_point = 0
- steepness = 0.5
- max_value = 1
- level_shift = 0

```python
sigmoid = Sigmoid(mid_point=5, steepness=0.05, max_value=1, level_shift=0)
```

#### Set number of input samples 

```python
number_of_samples = 100
```

#### Initialize input and output array 

```python
sigmoid_result = numpy.zeros( number_of_samples )
sigmoid_in = numpy.zeros( number_of_samples )

start_val = int(number_of_samples/2)
stop_val = int((-1*start_val)-1)
```

#### Generate Sigmoid 

```python
for i in range(start_val,stop_val,-1):
    sigmoid_result[i-start_val] = sigmoid.generate( i )
    sigmoid_in[i-start_val] = i
    
```

#### Plot 

```python
plt.title( "Sigmoid ($n = " + str( number_of_samples ) + "$ steps)" )
plt.xlabel("Input samples")
plt.ylabel("Output values")
plt.plot( sigmoid_in,sigmoid_result)
plt.show()
```

###  Personnel Walk angle characterstics

```python
mid_point = -2
steepness = 0.5
max_speed_mps = 10.0 # imagine Bolt is running here 100 meters in 9.58 seconds :D
```

```python
walk_angle_gen = WalkAngleGenerator( mid_point=mid_point, steepness=steepness, max_value=math.radians(135), level_shift=math.radians(45))
```

### Angle Vs Velocity (Walk-angle) Graph velocity from stand-still to 40 meter-per-second

```python
x_axis_start = 0
x_axis_stop = 20

walk_angle_result = numpy.zeros( x_axis_stop )
velocity_in = numpy.zeros( x_axis_stop )

prev_angle = 0.0
for i in range( x_axis_start, x_axis_stop ):
    walk_angle_result[i] = math.degrees(walk_angle_gen.get_angle_deviation(i))
    velocity_in[i] = i
    
plt.title( "Walk angle" )
plt.xlabel("Velocity")
plt.ylabel("Angle in degrees")
plt.plot( velocity_in,walk_angle_result )
plt.show()
```

### Using Walk-angle as variation (Guassian distribution) with reference angle (previous angle) of walk  being zero

```python
x_axis_start = 0
x_axis_stop = 40

walk_angle_result = numpy.zeros( x_axis_stop )
velocity_in = numpy.zeros( x_axis_stop )

prev_angle = 0.0
for i in range( x_axis_start, x_axis_stop ):
    walk_angle_result[i] = math.degrees(walk_angle_gen.generate(prev_angle,i))
    velocity_in[i] = i
    
plt.title( "Walk angle" )
plt.xlabel("Velocity")
plt.ylabel("Angle in degrees")
plt.plot( velocity_in,walk_angle_result )
plt.plot( velocity_in,numpy.full(shape=x_axis_stop, fill_value=0), linestyle="dashed", label="0 degree" )
plt.plot( velocity_in,numpy.full(shape=x_axis_stop, fill_value=45), linestyle="dashed", label="45 degree" )
plt.plot( velocity_in,numpy.full(shape=x_axis_stop, fill_value=-45), linestyle="dashed", label="-45 degree" )
plt.legend()
plt.show()
```

### Using Walk-angle as variation (Guassian distribution) to the accumulated angle (previous angle) of walk 

```python
x_axis_start = 0
x_axis_stop = 40

walk_angle_result = numpy.zeros( x_axis_stop )
walk_angle_without_variation = numpy.zeros( x_axis_stop )
velocity_in = numpy.zeros( x_axis_stop )

prev_angle = 0.0
for i in range( x_axis_start, x_axis_stop ):
    walk_angle_result[i] = math.degrees(walk_angle_gen.generate(prev_angle,i))
    walk_angle_without_variation[i] = math.degrees(walk_angle_gen.generate(0,i))
    velocity_in[i] = i
    prev_angle = math.radians(walk_angle_result[i])
    
plt.title( "Walk angle" )
plt.xlabel("Velocity")
plt.ylabel("Angle in degrees")
plt.plot( velocity_in,walk_angle_result, label="Walk with Variation" )
plt.plot( velocity_in,walk_angle_without_variation, label="Walk without Variation")
plt.legend()
plt.show()
```
