### Generate Surface example

#### Define parameters
- Dimension of flat-world (x,y)
- roughness of the surface

```python
area_dimension = (50, 50)
surface_roughness = 1
```

#### Create a Terrain object with above mentioned parameters

```python
Terrain_1 = TerrainGenerator( dimension=area_dimension, roughness=surface_roughness )
```

#### Generate Terrain

```python
Terrain_1.generate()
```

#### Plot the generated Terrain

```python
fig = plt.figure()
ax = plt.axes( projection='3d' )
ax.plot_wireframe( Terrain_1.x_axis, Terrain_1.y_axis, Terrain_1.z_axis, color='red' )
ax.set_title( 'wireframe' )
plt.show()
```
